```python
import pandas as pd
```

```python
metadata = pd.read_csv('/Users/harrisonma/Documents/PhD Classes/2025 Spring/CSE 283/CSE-283-Proj/data/pnas_patient_info.csv', index_col=0)
metadata.head()
```

```python
metadata_df = metadata[['datespecimen','On.Plate','er_cat','pr_cat','her2_cat','cancerstage_cat','chemo','datechemostart','datechemoend']]
metadata_df
```

```python
metadata_df['chemo_encoded'] = metadata_df['chemo'].map({
    'TC': 1,
    'AC/T': 2,
    'AC/T/D': 3
})
metadata_df.head()
```

```python
metadata_df['cancerstage_cat'].median()
# for metadata_df cancer stage > 5, set to 2
metadata_df['cancerstage_cat'] = metadata_df['cancerstage_cat'].replace(9, 2)
metadata_df['cancerstage_cat'].value_counts()
```

```python
metadata_df['cancer_subtype'] = metadata_df['er_cat'].astype(str) + '_' + metadata_df['pr_cat'].astype(str) + '_' + metadata_df['her2_cat'].astype(str)
metadata_df['cancer_subtype'].value_counts()
```

```python
# one-hot encode the cancer subtype
metadata_df = pd.get_dummies(metadata_df, columns=['cancer_subtype'], prefix='', prefix_sep='')
# drop the original categorical columns
metadata_df = metadata_df.drop(columns=['er_cat', 'pr_cat', 'her2_cat','chemo'])
# rename the columns to remove the 'cancerstage_cat_' prefix
metadata_df = metadata_df.rename(columns=lambda x: x.replace('cancerstage_cat_', ''))
# rename chemo_encoded to chemo
metadata_df = metadata_df.rename(columns={'chemo_encoded': 'chemo'})
metadata_df.head()
```

```python
# make a new column for chemo duration in days
metadata_df['datechemostart'] = pd.to_datetime(metadata_df['datechemostart'], errors='coerce',dayfirst=True)
metadata_df['datechemoend'] = pd.to_datetime(metadata_df['datechemoend'], errors='coerce',dayfirst=True)
metadata_df['chemo_duration'] = (metadata_df['datechemoend'] - metadata_df['datechemostart']).dt.days
# drop the date columns

metadata_df.head()
```

```python
metadata_df['batch'] = metadata_df['On.Plate'].astype(str).apply(lambda x:x[0])
metadata_df
```

```python
metadata_df['collection_date'] = pd.to_datetime(metadata_df['datespecimen'], errors='coerce', dayfirst=True)
metadata_df = metadata_df.drop(columns=['datespecimen', 'On.Plate'])
metadata_df['collection_to_chemo_start'] = (metadata_df['datechemostart'] - metadata_df['collection_date']).dt.days
metadata_df = metadata_df.drop(columns=['datechemostart', 'datechemoend','collection_date'])

metadata_df.head()
```

```python
metadata_df.columns = ['cancer_stage','chemo_type','Triple_Negative','HER2_positive','ER_positive','PR_negative','Triple_Positive','chemo_duration','batch','collection_to_chemo_start']
# save the processed metadata to a new CSV file
metadata_df.to_csv('/Users/harrisonma/Documents/PhD Classes/2025 Spring/CSE 283/CSE-283-Proj/data/pnas_patient_info_processed.csv', index=False)
```
