```python
tumor_top2000 = tumor[top_2000.index]
tumor_top2000.head()
```

```python
recurrance_status = pd.read_csv('/Users/harrisonma/Documents/PhD Classes/2025 Spring/CSE 283/CSE-283-Proj/data/pnas_patient_info.csv', index_col=0)#['recurStatus']

metadata['chemo'] = recurrance_status['chemo'].values
metadata = pd.get_dummies(
    metadata,
    columns=["chemo"],
    prefix="chemo",      # optional: this will name the new columns “chemo_AC”, “chemo_AC/T”, etc.
    drop_first=True     # keep all 4 columns; if you set drop_first=True you’d only get 3 columns
)
```

```python
tumor_top2000.reset_index(inplace=True)
metadata.reset_index(inplace=True)
```

```python
# one-hot-encode batch column 
# metadata = pd.get_dummies(metadata, columns=['batch'], prefix='batch', drop_first=True)
```

```python
X = pd.concat([tumor_top2000, metadata], axis=1)
y = recurrance_status['recurStatus'].values
```

```python
X.drop(columns=['chemo_type'], inplace=True)
```

```python
X.loc[67, 'collection_to_chemo_start'] = X['collection_to_chemo_start'].mean()
```

```python
X.columns
```

```python
# remove columns level_0 and index from X
X.drop(columns=['index'], inplace=True)
# drop columns that start with 'batch'
X = X.loc[:, ~X.columns.str.startswith('batch')]
```

```python
X['recurStatus'] = y
X.to_csv('/Users/harrisonma/Documents/PhD Classes/2025 Spring/CSE 283/CSE-283-Proj/Harrison/recurrence_hvg_input.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, accuracy_score

param_grid = {
    "n_estimators": [100, 200, 500],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", 0.2, 0.5],
    "bootstrap": [True, False],
}

param_grid_small = {
    "n_estimators": [100, 300],
    "max_depth": [None, 20],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "max_features": ["sqrt", 0.2],
    "bootstrap": [True],
}

rf = RandomForestClassifier(random_state=42, n_jobs=-1, class_weight="balanced")
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    scoring="accuracy",
    cv=cv_strategy,
    n_jobs=-1,
    verbose=2,
)

grid_search.fit(X, y)

print("Best parameters found:\n", grid_search.best_params_)
print("\nBest crossval score: {:.4f}".format(grid_search.best_score_))

best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X)
print("\nTraining accuracy:", accuracy_score(y, y_pred))
print("\nClassification report (on training data):\n", classification_report(y, y_pred))
```

```python
print("Best parameters found:\n", grid_search.best_params_)
print("\nBest crossval score: {:.4f}".format(grid_search.best_score_))

best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X)
print("\nTraining accuracy:", accuracy_score(y, y_pred))
print("\nClassification report (on training data):\n", classification_report(y, y_pred))
```

```python
# find features importance
importances = best_rf.feature_importances_
feature_names = X.columns
indices = np.argsort(importances)[::-1]
print("\nFeature importances (top 20):")
for f in range(20):
    print(f"{f + 1}. {feature_names[indices[f]]}: {importances[indices[f]]:.4f}")
```
